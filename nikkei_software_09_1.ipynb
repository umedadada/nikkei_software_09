```python
import pandas as pd
import plotly.express as px
```

```python
# データの読み込み
df=pd.read_csv('baseball_players.csv')
```

```python
# データのサイズの確認
df.shape
```

```python
# 変数の確認
df.info()
```

```python
# ヒストグラム
fig = px.histogram(df, x='height', nbins=20, barmode='overlay')
fig.show()
```

```python
# 平均値
sum(df['height'])/len(df['height'])
```

```python
# 中央値
sorted_list=sorted(df['height']) ##リストを昇順に並び替える
if len(sorted_list)%2==0: ##リストが偶数の場合
 #割り算の整数部（整数除算）://演算子
 median=(sorted_list[len(sorted_list)//2-1]+sorted_list[len(sorted_list)//2])/2
 print(median)
elif len(sorted_list)%2==1:#リストが奇数の場合
 median=sorted_list[len(sorted_list)//2+1]
 print(median)
```

```python
# 分散と標準偏差の計算
## 平均値の算出
avg=sum(df['height'])/len(df['height'])
# 平方和の初期値を設定する
s=0
# 平方和を計算する
for i in range(len(df['height'])):
  s+=(df['height'][i]-avg)**2
# 不偏分散の計算
var=s/(len(df['height']) - 1)
# 不偏標準偏差の計算
sigma=var**0.5

print(var)
print(sigma)
```

```python
# 日本人選手と海外選手のデータのばらつきの比較
## カラムでグループ化
grouped=df.groupby('JPNorNot')
## pandasの関数で計算
print(grouped['height'].var()) #分散
print(grouped['height'].std()) #標準偏差
```

```python
# ヒストグラム
fig = px.histogram(df, x="height", nbins=50, color="JPNorNot",barmode='overlay')
fig.show()
```

```python
# 箱ひげ図
fig = px.box(df, x="JPNorNot",y="height", color="JPNorNot")
fig.show()
```

```python
# 散布図
# fig = px.scatter(df, x="height", y="weight", color="JPNorNot")
fig = px.scatter(df, x="height", y="weight")
fig.show()
```

```python
# 相関係数の計算

## リストの取得
x=df['height']
y=df['weight']

## 平均値の算出
avg_x=sum(x)/len(x)
avg_y=sum(y)/len(y)

## 変数の初期化
sxx,syy,sxy=[0,0,0]

## 平方和の算出
for i in range(len(df)):
 sxx+=(x[i]-avg_x)**2
 syy+=(y[i]-avg_y)**2
 sxy+=(x[i]-avg_x)*(y[i]-avg_y)

# 相関係数の算出
r=(sxy/len(df))/((sxx/len(df))**0.5*(syy/len(df))**0.5)

print(r)
```

```python
# pandasの関数を使うと楽である
print(df['height'].mean()) #平均値
print(df['height'].median()) #中央値
print(df['height'].var()) #不偏分散の計算
print(df['height'].std()) #不偏標準偏差の計算
print(df[['height','weight']].corr().at['height','weight']) #相関係数
```
